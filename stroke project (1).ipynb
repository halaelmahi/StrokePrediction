spliting x and y into training and validation sets

```python
from imblearn.over_sampling import RandomOverSampler

X_resampled, y_resampled = RandomOverSampler().fit_resample(X, y)
data_resampled = X_resampled.copy()
data_resampled['stroke'] = y_resampled
sns.set(rc={'figure.figsize':(6,4)})
plt.title('Target: Stroke')
sns.countplot(x = 'stroke', data = data_resampled)
```

deals with the class imbalance

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score

 
weight_range = [int(x) for x in np.logspace(1, 2.5, num=5)] # Weights for imbalanced data
min_leaf_range = [int(x) for x in np.linspace(8, 20, num=5)]

best_score = 0
best_combination = (10, 5)

for w in weight_range:
    for leaf_s in min_leaf_range:
        rf_model = RandomForestClassifier(
            n_estimators=150, criterion='gini',
            class_weight={0: 1, 1: w}, min_samples_leaf=leaf_s,
            max_features=None
        )
        rf_model.fit(X_train, y_train)
        y_pred_val = rf_model.predict(X_val)
        current_score = f1_score(y_val, y_pred_val)
        print("Weight:", w, "\tMin leaf sample:", leaf_s, "\tF1-Score:", current_score)
        if current_score > best_score:
            best_score = current_score
            best_combination = (w, leaf_s)

w, leaf_s = best_combination
rf_model = RandomForestClassifier(
    n_estimators=150, criterion='gini',
    class_weight={0: 1, 1: w}, min_samples_leaf=leaf_s,
    max_features=None
)
rf_model.fit(X_train, y_train)
```

```python
y_pred_val = rf_model.predict(X_val)
val_score = f1_score(y_val, y_pred_val)
print("F1-Score on Validation Set:", val_score)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, precision_recall_fscore_support

 
 
param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
grid_search = GridSearchCV(SVC(), param_grid, scoring='f1')
grid_search.fit(X_train, y_train)

 
best_params = grid_search.best_params_

 
svm = SVC(C=best_params['C'], kernel=best_params['kernel'])
svm.fit(X_train, y_train)

 
y_pred = svm.predict(X_val)

 
precision, recall, f1_score, _ = precision_recall_fscore_support(y_val, y_pred, average='binary')
print("F1-Score on Validation Set (Support Vector Machines):", f1_score)
```
